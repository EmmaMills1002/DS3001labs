1. This paper focuses on an important component of data cleaning, data tidying, and goes on to emphasize the importance of tidying data into a uniform structure for further analysis.
2. The tidy data standard provides a method for organizing datasets to be more functional for analysis. Wickham has developed two packages, reshape and reshape2, to create a framework for this process.
3. Wickham uses the Leo Tolstoy quote about families to make the point that data can be messy in a wide variety of ways, which is often why it can be difficult to clean data in a standard way. He goes on to make the point that it is difficult to distinguish between what should be classified as an observation and what should be classified as a variable, offering the difference between "height and weight" and "height and width" to distinguish between what is a value and what is a variable.
4. Wickham describes a value as the quantitative or qualitative information found in a dataset. A variable contains all values that measure the same underlying attribute across units (such as height, temperature, or duration). An observation contains all values measured on the same unit across attributes (such as a person, day, or city).
5. "Tidy Data" is defined as a dataset in which each variable forms a column, each observation forms a row, and each type of observational unit forms a table.
6. The most common problems with messy datasets are
- Column headers are values, not variable names
- Multiple variables are stored in one column
- Variables are stored in both rows and columns
- Multiple types of observational units are stored in the same table
- A single observational unit is stored in multiple tables
The example shown in table four is messy because there are actually three variables represented, and values are found in the column names. Melting the data stacks it so that each column represents a variable and each row represents one specific observation.
7. Table 7 is messy because wk1, wk2, wk3... are actually values that are shown as columns. Table 8 is tidier because it has a "week" column, where the observation describes each week.
8. Wickham describes the "chicken and the egg" problem of data as the idea that tidy data's usefulness is based on the tools that work with it, and tidy tools are related to tidy data. He hopes that in the future, the tidy data framework is standardized and that people will continue to build on it to develop even better data storage strategies and tools.
